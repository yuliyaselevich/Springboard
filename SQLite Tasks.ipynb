```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members!

query_Q10 = '''SELECT *
FROM 
(SELECT c.facility, SUM(c.cost) AS total_revenue
FROM (SELECT b.memid, b.slots, b.facid, f.name AS facility, f.membercost, f.guestcost,
             CASE WHEN b.memid = 0.0 THEN f.guestcost*b.slots
                  ELSE f.membercost*b.slots
             END AS cost
      FROM Bookings AS b
      INNER JOIN Facilities AS f
      ON b.facid = f.facid) AS c
GROUP BY c.facility
ORDER BY SUM(c.cost) DESC) AS r
WHERE total_revenue < 1000
'''

fac_rev_df = pd.read_sql_query(query_Q10, engine)
fac_rev_df
```

```python
#Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order 

query_Q11 = '''SELECT m.firstname || ' ' || m.surname AS Member_name, r.firstname || ' ' || r.surname AS Recommended_by
FROM Members AS m
INNER JOIN Members AS r
ON m.recommendedby = r.memid
ORDER BY Member_name
'''
fac_rev_df = pd.read_sql_query(query_Q11, engine)
fac_rev_df
```

```python
#Q12: Find the facilities with their usage by member, but not guests 
query_Q12 = '''SELECT COUNT(b.bookid) AS used_by_members, f.name AS facility
FROM Bookings AS b
INNER JOIN Facilities AS f
ON b.facid = f.facid
WHERE b.memid <> 0
GROUP BY facility
ORDER BY facility
'''
fac_rev_df = pd.read_sql_query(query_Q12, engine)
fac_rev_df
```

```python
#Q13: Find the facilities usage by month, but not guests
query_Q13 = '''SELECT COUNT(b.bookid) AS used_by_members, f.name AS facility, strftime('%m', b.starttime) AS month
FROM Bookings AS b
INNER JOIN Facilities AS f
ON b.facid = f.facid
WHERE b.memid <> 0
GROUP BY facility, month
ORDER BY month, facility
'''
fac_rev_df = pd.read_sql_query(query_Q13, engine)
fac_rev_df
```
